```python
def retrieve_descriptions(iconclasses):
    description = []
    iconclasses = ast.literal_eval(iconclasses)
    for iclass in iconclasses:
        curr_class = labels[iclass]
        if(curr_class!=labels[-1] and curr_class!=labels[-2] and curr_class!=labels[-3]):
            response_m = requests.get("https://iconclass.org/{}.jsonld".format(curr_class))
            response_m = response_m.json()
            description.append(response_m['graph'][0]['prefLabel'][-1]['value'])
        elif(curr_class==labels[-1]):
            response_m = requests.get("https://iconclass.org/11D.jsonld".format(curr_class))
            response_m = response_m.json()
            description.append(response_m['graph'][0]['prefLabel'][-1]['value'])
        
        elif(curr_class==labels[-2]):
            response_m = requests.get("https://iconclass.org/11F4.jsonld".format(curr_class))
            response_m = response_m.json()
            description.append(response_m['graph'][0]['prefLabel'][-1]['value'])
        
        elif(curr_class==labels[-3]):
            response_m = requests.get("https://iconclass.org/11F.jsonld".format(curr_class))
            response_m = response_m.json()
            description.append(response_m['graph'][0]['prefLabel'][-1]['value'])
    
    return " ".join(description)
```
